```python
from tensorflow import keras
from keras.models import load_model
import pandas as pd
from keras import activations
from keras import applications
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from vis.visualization import visualize_activation
from vis.utils import utils
```

With the help of keras-vis (https://raghakot.github.io/keras-vis/) I want to analyse and see what the model has learned.

```python
# Load model here
model = load_model('models/first_model/model.h5')
```

```python
model.summary()
```

### Let's see which picture would generate the maximum activation of our last layer

```python
figure(num=None, figsize=(20, 6),dpi=80, facecolor='w', edgecolor='k')
layer_idx = utils.find_layer_idx(model,'dense_2')
model.layers[layer_idx].activation = activations.linear
model = utils.apply_modifications(model)
img3 = visualize_activation(model,layer_idx)
plt.imshow(img3)
```
